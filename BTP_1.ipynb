```python
data1 = d1.copy()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
ks_1=[2,4,6,8,10,12]
labels_1=[]
centroids_1=[]
sil_score_1=-2
K=0
inertias_1=[]
for k in ks_1:
    km=KMeans(n_clusters=k,random_state=42)
    lbl=km.fit_predict(data1)
    ctds=km.cluster_centers_
    inertia = km.inertia_
    silhouette = silhouette_score(data1, lbl)
    print("Inertia:", inertia, "Silhouette:", silhouette)
    inertias_1.append(inertia)
    if sil_score_1<silhouette:
        K=k
        sil_score_1=silhouette
        labels_1=lbl
        centroids_1=ctds
```

```python
# Plot inertia (elbow method)
import matplotlib.pyplot as plt

plt.plot(ks_1, inertias_1, marker='o')
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

# Our Solution : Fair K-Means Clustering

**Introducing fairness term**



**Continuous Sensitive Attribute : Age**

```python
# d1 is original dataset
```

```python
# This function regturns us windows (array of windows).
# width is W_percent*range, where range is (max-min)
# stride fraction is by how much our pointer will move to make next window

def make_windows(arr,W_percent=0.2,stride_frac=0.2):
    windows=[]
    mx=float(arr.max())
    mn=float(arr.min())
    R=mx-mn;
    width=W_percent*R
    actual_mn=mn-(width-stride_frac);   # we will be taking range as (actual_min, actual_mx) as by this each original range point will come in equal number of windows
    actual_mx=mx+(width-stride_frac);
    st=actual_mn
    while st+width<=actual_mx:
      windows.append((st,st+width))
      st+=stride_frac
    return windows
```

```python
# This is a helper function which returns a matrix of size (no of points,no of windows)
# This matrix tells for each datapoint in which window it falls, a point will fall in multiple windows as windows will be overlapping
# So for a point if that point is in some window that matrix entry will be 1 else 0
# age_arr : age column
# windows : windows array we got from makewindow function

def membership_matrix(age_arr, windows):
    N = len(age_arr)
    W = len(windows)
    mem = np.zeros((N, W), dtype=np.int32)
    for j, (l, r) in enumerate(windows):
      for i in range(N):
        age=age_arr[i]
        if(j==W-1):
          if ((age>=l) & (age<=r)):
              mem[i][j]=1
        else:
           if ((age>=l) & (age<r)):
              mem[i][j]=1
    return mem
```

```python
# This function takes input about point to window matrix and returns Hx : the total proportion of age in each window for whole dataset

def compute_hX(membership):
    N = membership.shape[0]
    return membership.sum(axis=0) / float(N)
```

```python
# This function takes inputs as:
# 1)cluster sizes: array telling size of each cluster. length k ints
# 2)Cluster_window count: matrix telling count of points in each window for each cluster . (k, m) ints  k clusters and m windows
# 3)Hx: Totol proportion of age in each window for whole dataset. hX: length m floats

# This function returns the fairness loss

def compute_JF_from_counts(cluster_window_counts, cluster_sizes, hX):
    k, m = cluster_window_counts.shape
    total = 0.0
    for c in range(k):
        sz = cluster_sizes[c]
        if sz <= 0:
            continue
        hC = cluster_window_counts[c] / float(sz)  # here cluster_window_counts[c] is array indicating counts of points in each window belonging to that cluster
        total += np.abs(hC - hX).sum()
    return total
```

```python
# This function returns SSE (sum squared error), Utitlity Loss
# inputs are Dataset, labels, centroids

def compute_JU(X, labels, centroids):
    diffs = X - centroids[labels]
    sse = np.sum(diffs * diffs)
    return float(sse)
```

```python
from sklearn.metrics import pairwise_distances
```

```python
# Simple fair k-means with pooling-window fairness term.
# major inputs:
# 1) X: data
# 2) k: no of cluster
# 3) lambdaa: fairness-utility tradeoff constanst
# 2) age_arr: age column (sensitive attribute)
# 3) W_percent: needed in make_windows function for taking width
# 4) stride_frac: needed in make_windows functions to know by how much to move forword for making windows
# 5) n_min: number of minimum poinnts we want to keep in a cluster
# 6) max_iter: Max number of iteration for EM step

# This function returns labels,centroids,windows,final_ju (utility_loss), final_jf (fairness_loss)


def fair_kmeans(X, age_arr, k=4, lambdaa=0.8, W_percent=0.2, stride_frac=0.2, n_min=1,
                max_iter=20, random_state=42, verbose=True):

    N, d = X.shape

    # 1) windows & membership
    windows = make_windows(age_arr, W_percent=W_percent, stride_frac=stride_frac)
    mem = membership_matrix(age_arr, windows)     # (N, m)
    hX = compute_hX(mem)                         # (m,)
    m = mem.shape[1]

    # Normalization constants according to our paper
    Tu = N * d
    Tf = k * m

    # 2) initialize with sklearn kmeans
    kmeans_init = KMeans(n_clusters=k, random_state=random_state, n_init=10).fit(X)
    labels = kmeans_init.labels_.copy()
    centroids = kmeans_init.cluster_centers_.copy()

    # 3) cluster meta info
    cluster_sizes = np.bincount(labels, minlength=k).astype(int)  #array telling size of each cluster. length k ints
    cluster_window_counts = np.zeros((k, m), dtype=int)    # matrix telling count of points in each window for each cluster . (k, m) ints  k clusters and m windows
    for i in range(N):
        cluster_window_counts[labels[i]] += mem[i]

    # Precompute distances to centroids at start. pairwise means from eavery point to every centroid so its (Number of points ,number of centroids)
    dists = pairwise_distances(X, centroids, metric='euclidean', squared=True)

    for it in range(max_iter):
        changed = 0  # track how many points changed clusters
        order = np.random.permutation(N) # randomize order for fairness in updates

        # E-STEP: Reassign each point to best cluster
        for i in order:
            cur = labels[i]  # current cluster assignment
            best_deltaJ = 0.0
            best_j = cur  # by default, stay in same cluster

            # Precompute fairness sums for the current cluster
            sz_cur = cluster_sizes[cur]
            prev_sum_cur = 0.0 if sz_cur == 0 else np.abs(cluster_window_counts[cur] / float(sz_cur) - hX).sum()

            # Try moving this point to every other cluster
            for j in range(k):
                if j == cur:
                    continue # skip same cluster
                if cluster_sizes[cur] - 1 < n_min:
                    continue  # don't allow empty/small clusters

                # Calculate Delta Ju (Utility Change)
                # JU: Change in utility term
                # If moved to cluster j, how much does total distance change?
                deltaU = (dists[i, j] - dists[i, cur]) / Tu

                # Calculate Delta Jf (Fairness Change)

                # Compute fairness deviation before move
                sz_j = cluster_sizes[j]
                prev_sum_j = 0.0 if sz_j == 0 else np.abs(cluster_window_counts[j] / float(sz_j) - hX).sum()

                # New counts and sizes after moving point i
                cnt_cur_new = cluster_window_counts[cur] - mem[i]
                cnt_j_new = cluster_window_counts[j] + mem[i]
                sz_cur_new = sz_cur - 1
                sz_j_new = sz_j + 1

                # Compute new fairness deviations after the move
                new_sum_cur = 0.0 if sz_cur_new == 0 else np.abs(cnt_cur_new / float(sz_cur_new) - hX).sum()
                new_sum_j = 0.0 if sz_j_new == 0 else np.abs(cnt_j_new / float(sz_j_new) - hX).sum()

                # Change in fairness term (scaled)
                deltaF = ((new_sum_cur + new_sum_j) - (prev_sum_cur + prev_sum_j)) / Tf

                # Total Objective Change
                # Overall objective = (1 - )*JU + *JF
                deltaJ = (1 - lambdaa) * deltaU + lambdaa * deltaF

                # If total objective decreases (better), keep this move
                if deltaJ < best_deltaJ:
                    best_deltaJ = deltaJ
                    best_j = j

            # If an improving move was found, update the assignments
            if best_j != cur:
                labels[i] = best_j
                # Update cluster statistics accordingly
                cluster_window_counts[cur] -= mem[i]
                cluster_window_counts[best_j] += mem[i]
                cluster_sizes[cur] -= 1
                cluster_sizes[best_j] += 1
                changed += 1

        # M-step: Update cluster centroids based on new assignments
        for c in range(k):
            idx = np.where(labels == c)[0]
            if len(idx) > 0:
                centroids[c] = X[idx].mean(axis=0)

        # Recompute distances to new centroids for the next iteration
        dists = pairwise_distances(X, centroids, metric='euclidean', squared=True)

        if verbose:
            print(f"Iteration {it+1}/{max_iter}, Points changed: {changed}")

        # If no points changed cluster  convergence reached
        if changed == 0:
            break

    # 7) FINAL METRICS (Utility and Fairness)
    final_ju_scaled = compute_JU(X, labels, centroids)/Tu
    final_jf_scaled = compute_JF_from_counts(cluster_window_counts, cluster_sizes, hX)/Tf

    if verbose:
        print(f"Finished. Final JU: {final_ju_scaled}, Final JF: {final_jf_scaled}")

    return labels, centroids, windows, final_ju_scaled, final_jf_scaled
```

### Tradeoff between utility loss and fairness loss

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances

lambda_values = np.linspace(0, 1, 11)  # Creates 11 points: 0, 0.1, 0.2, ..., 1.0

# Lists to store the results
utility_losses = []
fairness_losses = []

# --- Run the Loop ---
for lam in lambda_values:
    print(f"----- Running Fair K-Means for lambda = {lam:.1f} -----")

    # Run the algorithm
    _, _, _, ju, jf = fair_kmeans(
        data3.values,
        age_std,
        k=6,
        lambdaa=lam,
        n_min=50,
        max_iter=10,  # Keep max_iter low for quick testing, increase for final results
        random_state=42,
        verbose=False # Set to False to keep the output clean for the loop
    )

    # Store the results
    utility_losses.append(ju)
    fairness_losses.append(jf)

print("\\n----- Experiment Complete -----")
```

```python
# Using Random Search for hyperparametertuning: taking inputs as data and number of iterations
# in each iteration taking random lambdaa, stride_fraction, number of cluster

def random_search_fiar_kmeans(X,age_arr,n_iter=30):
  best_score=float('inf')
  best_params=None
  best_result=None

  for t in range(n_iter):
    lambdaa=np.random.uniform(0.1,1.0)
    W_percent=np.random.uniform(0.1,1.0)
    stride_frac=np.random.uniform(0.1,1.0)
    k=np.random.choice([3,4,5,6,7,8,9])

    labels, centroids, windows, JU, JF = fair_kmeans(
            X, age_arr, k=k, lambdaa=lambdaa,
            W_percent=W_percent, stride_frac=stride_frac, verbose=False
        )

    total_loss=(1-lambdaa)*(JU) + (lambdaa)*(JF)

    print(f"Trial {t+1}: ={lambdaa:.2f}, W={W_percent:.2f}, stride={stride_frac:.2f}, k={k}, loss={total_loss:.4f}")

    if total_loss < best_score:
        best_score = total_loss
        best_params = (lambdaa, W_percent, stride_frac, k)
        best_result = (labels, centroids, JU, JF)

  print(" Best configuration:")
  print(f"={best_params[0]:.2f}, W={best_params[1]:.2f}, stride={best_params[2]:.2f}, k={best_params[3]}")
  print(f"Final loss={best_score:.4f}")

  return best_params, best_result

```
